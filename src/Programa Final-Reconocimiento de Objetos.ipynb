```python
##Programa de reconocimiento de objetos
##Elaborado por: Eric Alejandro Song Laguna

## Bibliotecas a usar para este programa ##
import cv2
import os
import imutils

## Esta parte del programa creara una carpeta donde guardaremos los resultados positivos y sera la primera opcion
## del menu

def capturar_imagenes(Datos):
    if not os.path.exists(Datos):
        print('Carpeta creada:', Datos)
        os.makedirs(Datos)
        
## A partir de aqui se iniciara el programa para la captura usando la biblioteca cv2, el count hara que
## se genere un contador para las imagenes tomadas        
    
    cap = cv2.VideoCapture(0)
    x1, y1 = 190, 80
    x2, y2 = 450, 398
    count = 0
    
## Aqui se creara un rectangulo que delimitara el area de nuestro modelo para obtener los resultados    
    
    while True:
        ret, frame = cap.read()
        if ret == False:
            break
        
        imAux = frame.copy()
        cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
        objeto = imAux[y1:y2, x1:x2]
        objeto = imutils.resize(objeto, width=38)
        
## Esta parte guardara la imagen que tomemos una vez que la tengamos frente a la camara, se guardara con
## la tecla "s"        
        
        k = cv2.waitKey(1)
        if k == ord('s'):
            cv2.imwrite(Datos + '/objeto_{}.jpg'.format(count), objeto)
            print('Imagen guardada:' + '/objeto_{}.jpg'.format(count))
            count += 1
        if k == 27:
            break
        
        cv2.imshow('frame', frame)
        cv2.imshow('objeto', objeto)
        
        key = cv2.waitKey(1)
        if key == 27:  # Si se deja presionada la tecla ESC se cierra el programa
            break
    
    cap.release()
    cv2.destroyAllWindows()

## Esta parte del programa creara una carpeta donde guardaremos los resultados negativos y sera la segunda opcion
## del menu    

def capturar_imagenes_negativas(Datos2):
    if not os.path.exists(Datos2):
        print('Carpeta creada:', Datos2)
        os.makedirs(Datos2)
    
# A partir de aqui se iniciara el programa para la captura usando la biblioteca cv2, el count hara que
## se genere un contador para las imagenes tomadas 

    cap = cv2.VideoCapture(0)
    x1, y1 = 190, 80
    x2, y2 = 450, 398
    count = 0

## Aqui se creara un rectangulo que delimitara el area de nuestro modelo para obtener los resultados      
    
    while True:
        ret, frame = cap.read()
        if ret == False:
            break
        
        imAux = frame.copy()
        cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
        objeto = imAux[y1:y2, x1:x2]
        objeto = imutils.resize(objeto, width=38)
        
## Esta parte guardara la imagen que tomemos una vez que la tengamos frente a la camara, se guardara con
## la tecla "s"         
        
        k = cv2.waitKey(1)
        if k == ord('s'):
            cv2.imwrite(Datos2 + '/objeto_{}.jpg'.format(count), objeto)
            print('Imagen guardada:' + '/objeto_{}.jpg'.format(count))
            count += 1
        if k == 27:
            break
        
        cv2.imshow('frame', frame)
        cv2.imshow('objeto', objeto)
        
        key = cv2.waitKey(1)
        if key == 27:  # Si se deja presionada la tecla ESC se cierra el programa 
            break
    
    
    cap.release()
    cv2.destroyAllWindows()    
    
 #########################################################################################################
info_objetos = {
    "Disco DDH": "Este es un disco DDH utilizado en la detección.",
    "Otro Objeto": "Esta es información sobre otro objeto."
    # En esta parte se puede agregar informacion del objeto detectado
}
    
    
## Aqui se usara la linea para el uso de la camara en la deteccion    

def deteccion_objetos():
    cap = cv2.VideoCapture(0)
    Clasificador = cv2.CascadeClassifier('cascade.xml')   #llamamos el archivo de entrenamiento
## El programa que se utilizo para entrenar el archivo cascade genero este archivo a partir de las fotos tomadas

    while True:
        ret, frame = cap.read()
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        toy = Clasificador.detectMultiScale(gray, scaleFactor=10, minNeighbors=350) 

#Se cambia a escala de grises las fotos para poder ser utilizadas

        for (x, y, w, h) in toy:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
            objeto_detectado = "Disco DDH"  # Objeto a detectar el cual mostrara informacion

            if objeto_detectado in info_objetos:
                informacion = info_objetos[objeto_detectado]
                cv2.putText(frame, informacion, (x, y - 30), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        cv2.imshow('Captura', frame)

        key = cv2.waitKey(1)
        if key == 27:  # Si se deja presionada la tecla ESC se cierra el programa
            break

    cap.release()
    cv2.destroyAllWindows()
    
## Aqui inicia el menu de opciones    

def opcion1():
    print("Ejecutando programa para capturar imágenes positivas...")
    capturar_imagenes('p')

def opcion2():
    Datos2 = 'n'
    print("Ejecutando segundo programa para capturar imágenes negativas...")
    capturar_imagenes(Datos2)

def opcion3():
    print("Ejecutando programa de detección de objetos...")
    deteccion_objetos()

def main():
    while True:
        print("Bienvenido, este es el menú de opciones:")
        print("1. Capturar imágenes positivas")
        print("2. Capturar imagenes negativas")
        print("3. Ejecutar programa de detección de objetos")
        print("4. Salir")
    
        opcion = input("Seleccione una opción (1/2/3/4): ")
    
        if opcion == '1':
            opcion1()
        elif opcion == '2':
            opcion2()
        elif opcion == '3':
            opcion3()
        elif opcion == '4':
            print("Saliendo del programa. ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Por favor, seleccione una opción válida.")

if __name__ == "__main__":
    main()
```
